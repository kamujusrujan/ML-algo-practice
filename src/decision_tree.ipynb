```python
# linear regression model 
from abc import ABC, abstractmethod

class Model(ABC) : 

    @abstractmethod
    def train(self, X , Y) : pass

    @abstractmethod
    def infer(self, X) : pass
```

```python
class DecisionTree(Model) : 
    
    def __init__(self):
        super().__init__()

    def train(self,X,Y) : 
        pass

    def infer(self, X):
        pass

    # def gini_index(self, X, Y, feature_index):
    #     N = X.shape[0] 
    #     split_feature = X[:,feature_index]
    #     possible_values = np.unique(split_feature)
    #     total_prob = 0
    #     for value in possible_values : 
    #         mask = split_feature==value
    #         prob_value = Y[mask].shape[0] / N
    #         total_prob += prob_value ** 2
    #     return 1- total_prob        

    def split_node(self,X,Y, feature_index ) : 
        N = X.shape[0]
        split_feature = X[:,feature_index]
        unique_splits = np.unique(split_feature)
        for split_criteria in unique_splits : 
            mask = split_feature == split_criteria 
            split_classes = Y[mask]
            self.gini_index(split_classes)

    def gini_index(self,Y ) : 
        N = Y.shape[0]
        y_classes = np.unique(Y)
        total_prob = 0
        for value in y_classes : 
            mask = Y==value
            prob_value = Y[mask].shape[0] / N
            total_prob += prob_value ** 2 
        return 1- total_prob
```
